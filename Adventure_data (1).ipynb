```python
import pandas as pd

files = {
    "sales_new": r"C:\Users\Bhumika\OneDrive\Fact_Internet_Sales_New.csv",
    "internet_sales": r"C:\Users\Bhumika\OneDrive\FactInternetSales.csv",
    "territory": r"C:\Users\Bhumika\OneDrive\DimSalesterritory.csv",
    "subcategory": r"C:\Users\Bhumika\OneDrive\DimProductSubCategory.csv",
    "category": r"C:\Users\Bhumika\OneDrive\DimProductCategory.csv",
    "customer": r"C:\Users\Bhumika\OneDrive\Dimcustomer.csv",
    "date": r"C:\Users\Bhumika\OneDrive\DimDate.csv",
    "product": r"C:\Users\Bhumika\OneDrive\DimProduct.csv"
}

dataframes = {}

for name, path in files.items():
    dataframes[name] = pd.read_csv(path)

print(dataframes.keys())
```

```python
for name, df in dataframes.items():
    dataframes[name] = df.drop_duplicates()
```

```python
for name, df in dataframes.items():
    print(f"\n{name}")
    print(df.dtypes)
```

```python
sales = dataframes['internet_sales']

sales['OrderDate'] = pd.to_datetime(sales['OrderDate'])
sales['DueDate'] = pd.to_datetime(sales['DueDate'])
sales['ShipDate'] = pd.to_datetime(sales['ShipDate'])
```

```python
sales['Profit'] = (
    sales['SalesAmount'] - sales['TotalProductCost']
)
```

```python
sales['SalesAmount'].sum()
```

```python
sales['Profit'].sum()
```

```python
top_products = (
    sales.groupby('ProductKey')['SalesAmount']
         .sum()
         .sort_values(ascending=False)
         .head(10)
)
print(top_products)
```
